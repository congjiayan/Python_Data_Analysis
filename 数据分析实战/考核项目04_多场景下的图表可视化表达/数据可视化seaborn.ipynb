```python
# 创建正弦函数及图表

def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1, 7):
        plt.figure(num=1,figsize = (10,4))
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)
sinplot()
# 创建正弦函数，及出图      
```

```python
# 1、set() 
# 简单切换为seaborn图表风格

sns.set()
sinplot(flip=10)
```

```python
# 2、set_style()
# 切换seaborn图表风格
# 风格选择包括："white", "dark", "whitegrid", "darkgrid", "ticks"

fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(2,1,1)
sns.set_style('whitegrid')
data = np.random.normal(size=(20,6)) + np.arange(6)/2
sns.boxplot(data = data)
plt.title('style--whitegrid')
# 仍然可以使用matplotlib的参数

ax2 = fig.add_subplot(2,1,2)
sns.set_style('darkgrid')
sinplot()
```

```python
# 3、despine()
# 设置图表坐标轴
# seaborn.despine(fig=None, ax=None, top=True, right=True, left=False, 
# bottom=False, offset=None, trim=False)

sns.set_style('ticks')

fig = plt.figure(figsize=(10,15))
plt.subplots_adjust(hspace=0.3)
# 创建图表

ax1 = fig.add_subplot(3,1,1)
sinplot()
sns.despine()
# 删除了上、右坐标轴

ax2 = fig.add_subplot(3,1,2)
sns.violinplot(data=data)
#sns.despine(offset=10, trim=True)
# offset：与坐标轴之间的偏移
# trim：为True时，将坐标轴限制在数据最大最小值

ax3 = fig.add_subplot(3,1,3)
sns.boxplot(data=data,palette='deep')
#sns.despine(left=True, right = False)
# top, right, left, bottom：布尔型，为True时不显示
```

```python
# 4、axes_style()
# 设置局部图表风格，可学习和with配合的用法

with sns.axes_style('darkgrid'):
    plt.subplot(2,1,1)
    sinplot()
# 设置局部图表风格，用with做代码块区分

sns.set_style('whitegrid')
plt.subplot(2,1,2)
sinplot()
# 外部表格风格
```

```python
# 5、set_context()
# 设置显示比例尺度
# 选择包括：'paper', 'notebook', 'talk', 'poster'

sns.set_context('paper')
sinplot()
# 默认为notebook
```

```python
sns.set_style("darkgrid")
sns.set_context("paper")
# 设置风格、尺度
```

```python
# 1、直方图 - distplot()

rs = np.random.RandomState(10)
s = pd.Series(rs.randn(100)*100)
plt.figure(figsize=(14,8))
sns.distplot(s,bins=10,hist=True,kde=True,norm_hist=True,rug=True,vertical=False,color='red',label='distplot',axlabel='x')
plt.legend(loc = 'best',fontsize = 14)

# bins → 箱数
# hist、kde → 是否显示箱/密度曲线
# norm_hist → 直方图是否按照密度来显示
# rug → 是否显示数据分布情况
# vertical → 是否水平显示
# color → 设置颜色
# label → 图例
# axlabel → x轴标注
```

```python
# 1、直方图 - distplot()
# 颜色详细设置
plt.figure(figsize=(8,6))
sns.distplot(s,rug=True,
             rug_kws={'color':'g'},    # 设置数据频率分布颜色
            kde_kws={'color':'k','lw':1,'label':'KDE','linestyle':'--'},# 设置密度曲线颜色，线宽，标注、线形
            hist_kws={'histtype':'step','lw':1,'alpha':1,'color':'g'},  # 设置箱子的风格、线宽、透明度、颜色
            )   
            # 风格包括：'bar', 'barstacked', 'step', 'stepfilled'
```

```python
# 2、密度图 - kdeplot()
# 单个样本数据密度分布图

plt.figure(figsize=(10,8))
sns.kdeplot(s,shade = False,   # 是否填充
           color = 'red',
           vertical=False)
sns.kdeplot(s,bw=5, label="bw: 0.2",
            linestyle = '-',linewidth = 1.2,alpha = 0.5)
sns.kdeplot(s,bw=20, label="bw: 2",
            linestyle = '-',linewidth = 1.2,alpha = 0.5)
# bw → 控制拟合的程度，类似直方图的箱数

sns.rugplot(s,height=0.1,color='k',alpha = 0.5)
# 数据频率分布图
```

```python
# 2、密度图 - kdeplot()
# 两个样本数据密度分布图

rs = np.random.RandomState(2)  # 设定随机数种子
df = pd.DataFrame(rs.randn(100,2),
                 columns = ['A','B'])
sns.kdeplot(df['A'],df['B'],
           cbar = True,    # 是否显示颜色图例
           shade = True,   # 是否填充
           cmap = 'Reds',  # 设置调色盘
           shade_lowest=False,  # 最外围颜色是否显示
           n_levels = 10   # 曲线个数（如果非常多，则会越平滑）
           )
# 两个维度数据生成曲线密度图，以颜色作为密度衰减显示

sns.rugplot(df['A'], color="g", axis='x',alpha = 0.5)
sns.rugplot(df['B'], color="r", axis='y',alpha = 0.5)
# 注意设置x，y轴
```

```python
# 2、密度图 - kdeplot()
# 两个样本数据密度分布图
# 多个密度图

rs1 = np.random.RandomState(2)  
rs2 = np.random.RandomState(5)  
df1 = pd.DataFrame(rs1.randn(100,2)+2,columns = ['A','B'])
df2 = pd.DataFrame(rs2.randn(100,2)-2,columns = ['A','B'])
# 创建数据

sns.kdeplot(df1['A'],df1['B'],cmap = 'Reds',
            shade = True,shade_lowest=False)
sns.kdeplot(df2['A'],df2['B'],cmap = 'Greens',
            shade = True,shade_lowest=False)
# 创建图表
```

```python
# 1、综合散点图 - jointplot()
# 散点图 + 分布图

rs = np.random.RandomState(2)  
df = pd.DataFrame(rs.randn(200,2),columns = ['A','B'])
# 创建数据


sns.jointplot(x = df['A'],y = df['B'],
             data = df,color = 'k',
             s = 50,edgecolor = 'w',linewidth = 1,    # 设置散点大小、边缘线颜色及宽度(只针对scatter）
             kind = 'scatter',      # 设置类型：“scatter”、“reg”、“resid”、“kde”、“hex”
             space = 0.2,    # 设置散点图和布局图的间距
             height = 10,   # 图表大小（自动调整为正方形）
             ratio=5,    # 图表大小（自动调整为正方形）
             marginal_kws=dict(bins=15,rug = True))     # 设置柱状图箱数，是否设置rug
```

```python
# 1、综合散点图 - jointplot()
# 散点图 + 分布图
# 六边形图

df = pd.DataFrame(rs.randn(500,2),columns = ['A','B'])
# 创建数据

with sns.axes_style("white"):
    sns.jointplot(x=df['A'], y=df['B'],data = df, kind="hex", color="r",height = 8,
                 marginal_kws=dict(bins=20))
```

```python
# 1、综合散点图 - jointplot()
# 散点图 + 分布图
# 密度图

rs = np.random.RandomState(15)
df = pd.DataFrame(rs.randn(300,2),columns = ['A','B'])
# 创建数据

g = sns.jointplot(x=df['A'], y=df['B'],data = df,
                  kind="kde", color="r",height = 8,
                  shade_lowest=False)
# 创建密度图

g.plot_joint(plt.scatter,c="k", s=30, linewidth=1, marker="*")
# 添加散点图
```

```python
# 1、综合散点图 - JointGrid()
# 可拆分绘制的散点图
# plot_joint() + ax_marg_x.hist() + ax_marg_y.hist()

sns.set_style('white')

tips = sns.load_dataset('tips')
print(tips.head())
# 导入数据

g = sns.JointGrid(x='total_bill',y='tip',data=tips,height = 10)
# 创建一个绘图表格区域，设置好x、y对应数据

g.plot_joint(plt.scatter,c = 'm',s = 50,edgecolor = 'white')    # 设置框内图表，scatter
g.ax_marg_x.hist(tips['total_bill'],color = 'g',alpha = 0.8,bins = np.arange(0,60,3))     # 设置x轴直方图，注意bins是数组
g.ax_marg_y.hist(tips['tip'],color = 'b',alpha = 0.8,orientation = 'horizontal',bins = np.arange(0,12,1))

g.annotate(stats.pearsonr,fontsize = 14)
# 设置标注，可以为pearsonr，spearmanr

plt.grid(linestyle = '--')
```

```python
# 1、综合散点图 - JointGrid()
# 可拆分绘制的散点图
# plot_joint() + plot_marginals()

g = sns.JointGrid(x='total_bill',y='tip',data = tips,height = 10)
# 创建一个绘图表格区域，设置好x、y对应数据

g = g.plot_joint(plt.scatter,s = 50,edgecolor = 'white',c = 'r')
plt.grid(linestyle = '--')

g.plot_marginals(sns.distplot,kde = True,color = 'r')    # 绘制x，y轴直方图

g.annotate(stats.pearsonr,fontsize = 16)
```

```python
# 1、综合散点图 - JointGrid()
# 可拆分绘制的散点图
# plot_joint() + plot_marginals()
# kde - 密度图

g = sns.JointGrid(x='total_bill',y='tip',data=tips,height = 10)
# 创建一个绘图表格区域，设置好x、y对应数据

g = g.plot_joint(sns.kdeplot,cmap = 'Reds_r')
plt.grid(linestyle = '--')

g.plot_marginals(sns.kdeplot,shade = True,color = 'r')     # 绘制x，y轴密度图
g.annotate(stats.pearsonr,fontsize = 14)
```

```python
# 2、矩阵散点图 - pairplot()

sns.set_style('white')

iris = sns.load_dataset('iris')
print(iris.head())

sns.pairplot(iris,
            kind = 'scatter',  # 散点图/回归分布图 {‘scatter’, ‘reg’}  
            diag_kind="hist",  # 直方图/密度图 {‘hist’, ‘kde’}
            hue="species",   # 按照某一字段进行分类
            palette="husl",  # 设置调色板
            markers=["o", "s", "D"],  # 设置不同系列的点样式（这里根据参考分类个数）
            height = 4,   # 图表大小
            )
```

```python
# 2、矩阵散点图 - pairplot()
# 只提取局部变量进行对比

sns.pairplot(iris,vars=["sepal_width", "sepal_length"],
             kind = 'reg', diag_kind="kde", 
             hue="species", palette="husl",height = 4)
```

```python
# 2、矩阵散点图 - pairplot()
# 其他参数设置

sns.pairplot(iris, diag_kind="kde", markers="o",
             plot_kws=dict(s=50,color='blue', edgecolor="g", linewidth=1),
             # 设置点样式
             diag_kws=dict(shade=True,color='g'),
             # 设置密度图样式
             height = 3
            )
```

```python
# 2、矩阵散点图 - PairGrid()
# 可拆分绘制的散点图
# map_diag() + map_offdiag()

g = sns.PairGrid(iris,hue="species",palette = 'hls',
                vars = ['sepal_length','sepal_width','petal_length','petal_width'],  # 可筛选
                height = 3)
# 创建一个绘图表格区域，设置好x、y对应数据，按照species分类

g.map_diag(plt.hist,
          histtype = 'barstacked',      # 可选：'bar', 'barstacked', 'step', 'stepfilled'
          linewidth = 1,edgecolor = 'w')
# 对角线图表，plt.hist/sns.kdeplot

g.map_offdiag(plt.scatter,edgecolor = 'w',linewidth = 1,s = 40)     # 设置点颜色、大小、描边宽度
# 其他图表，plt.scatter/plt.bar...

g.add_legend(fontsize = 14)
```

```python
# 2、矩阵散点图 - PairGrid()
# 可拆分绘制的散点图
# map_diag() + map_lower() + map_upper()

g = sns.pairplot(iris,height=3)
g.map_diag(sns.kdeplot,lw = 3)    # 设置对角线图表
g.map_upper(plt.scatter, color = 'r')     # 设置对角线上端图表
g.map_lower(sns.kdeplot,cmap = 'Blues_d')    # 设置对角线下端图表
```

```python
# 1、stripplot()
# 按照不同类别对样本数据进行分布散点图绘制

tips = sns.load_dataset('tips')
print(tips.head())

plt.figure(figsize = (10,8))
sns.stripplot(x="day",          # x → 设置分组统计字段
              y="total_bill",   # y → 数据分布统计字段
              # 这里xy数据对调，将会使得散点图横向分布
              data=tips,        # data → 对应数据
              jitter = 0.1,    # jitter → 当点数据重合较多时，用该参数做一些调整，也可以设置间距如：jitter = 0.1
              size = 15, edgecolor = 'w',linewidth=1,marker = 'o',  # 设置点的大小、描边颜色或宽度、点样式
              )
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('day',fontsize = 18)
plt.ylabel('total bill',fontsize = 18)
plt.title('stripplot\n',fontsize = 20)
```

```python
# 2、热图 - heatmap()
# 绘制半边热图

sns.set(style="white")
# 设置风格

rs = np.random.RandomState(33)
d = pd.DataFrame(rs.normal(size=(100, 26)))
corr = d.corr()   # 求解相关性矩阵表格
# 创建数据

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# 设置一个“上三角形”蒙版

cmap = sns.diverging_palette(220, 10, as_cmap=True)
# 设置调色盘

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=0.2)
# 生成半边热图
```
